```python
from sklearn.cluster import KMeans
x1=df.iloc[:,3:5]
print(x1)
inertia=[]
for n in range (1,11):
    algo = KMeans(n_clusters =  n,init='k-means++', n_init = 10 ,max_iter=300, 
                        tol=0.0001,  random_state= 111  , algorithm='elkan') 
    algo.fit(x1)
    algo.inertia_
    inertia.append(algo.inertia_)
```

```python
plt.figure(1 , figsize = (16 ,8))
plt.plot(np.arange(1 , 11) , inertia , 'o')
plt.plot(np.arange(1 , 11) , inertia , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , 
plt.ylabel('Inertia')
plt.show()
```

```python
#so we can take 5 clusters as input on taking anual income and spending score#
x2=df.iloc[:,2:5]
print(x2)
inertia2=[]
for n in range(1,15):
    algo1=KMeans(n_clusters=n)
    algo1.fit(x2)
    inertia2.append(algo1.inertia_)
    
```

```python
plt.figure(2,figsize=(15,5))
plt.plot(np.arange(1,15),inertia2,'o')
plt.plot(np.arange(1,15),inertia2,'-',alpha=.3)
plt.xlabel("clusert")
plt.ylabel("inertia")
plt.show
```
